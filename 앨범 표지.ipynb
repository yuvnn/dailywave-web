```python
import pandas as pd
import requests
from pandas import DataFrame
from bs4 import BeautifulSoup
from datetime import datetime
import os

melon ='https://www.melon.com/chart/month/index.htm?classCd=GN0800'
agent = 'Mozila/2.0'
file_path = 'melon_songs.csv'
df = pd.read_csv(file_path)
df=df['song_no']
```

```python
def get_image(song_no):
    song_url = f'https://www.melon.com/song/detail.htm?songId={song_no}'
    
    image_save_path = "./images"  # 저장할 디렉터리
    os.makedirs(image_save_path, exist_ok=True)  # 디렉터리 생성

    # 곡 상세 페이지 요청
    resp = requests.get(song_url, headers={'User-agent': agent})
    song_soup = BeautifulSoup(resp.text, 'html.parser')
    
    # 앨범 이미지
    image_div = song_soup.find('div', id='d_song_org')
    if image_div:
        image_tag = image_div.find('img', {'onerror': 'WEBPOCIMG.defaultAlbumImg(this);'})
        if image_tag and 'src' in image_tag.attrs:
            image_url = image_tag['src']  # 이미지 URL 추출
            try:
                # 이미지 다운로드
                img_resp = requests.get(image_url, stream=True)
                if img_resp.status_code == 200:
                    # 파일 저장 경로 생성
                    image_file_path = os.path.join(image_save_path, f"{song_no}.jpg")
                    with open(image_file_path, 'wb') as img_file:
                        img_file.write(img_resp.content)  # 이미지 저장
                    return image_file_path  # 파일 경로 반환
                else:
                    return "이미지 다운로드 실패"
            except Exception as e:
                return f"이미지 저장 중 오류: {e}"
        else:
            return "이미지 태그 없음"
    else:
        return "d_album_org div 없음"
    
for i in df:
    print(get_image(i))

```
